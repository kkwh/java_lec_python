# 성별에 따른 tips 차이?

* 남성 손님들의 tip의 평균, 여성 손님들의 tip의 평균
* 성별 tip의 최댓값
* 성별 tip의 최솟값

```python
# 성별에 따른 팁(tip)의 평균 계산
tip_mean_by_gender = tips.groupby('sex')['tip'].mean()

# 성별에 따른 팁(tip)의 최댓값 계산
tip_max_by_gender = tips.groupby('sex')['tip'].max()

# 성별에 따른 팁(tip)의 최솟값 계산
tip_min_by_gender = tips.groupby('sex')['tip'].min()

# 결과 출력
print("남성 손님들의 팁(tip) 평균:", tip_mean_by_gender['Male'])
print("여성 손님들의 팁(tip) 평균:", tip_mean_by_gender['Female'])
print("성별 팁(tip) 최댓값:")
print(tip_max_by_gender)
print("성별 팁(tip) 최솟값:")
print(tip_min_by_gender)
```

# 흡연 여부에 따른 tip의 차이?

* 흡연 여부(Yes/No) 별 tip의 평균

```python
tip_mean_by_smoker = tips.groupby('smoker')['tip'].mean()

print("흡연자의 팁 평균:", tip_mean_by_smoker['Yes'])
print("비흡연자의 팁 평균:", tip_mean_by_smoker['No'])
```

# 요일에 따른 tip의 차이?

* 요일 별 tip의 평균

```python
tip_mean_by_day = tips.groupby('day')['tip'].mean()

print("일요일 팁 평균:", tip_mean_by_day['Sun'])
print("토요일 팁 평균:", tip_mean_by_day['Sat'])
print("금요일 팁 평균:", tip_mean_by_day['Fri'])
print("목요일 팁 평균:", tip_mean_by_day['Thur'])
```

# 시간에 따른 tip의 차이?

* 시간 별 tip의 평균

```python
tip_mean_by_time = tips.groupby('time')['tip'].mean()

print("저녁 팁 평균:", tip_mean_by_time['Dinner'])
print("토요일 팁 평균:", tip_mean_by_time['Lunch'])
```

```python
# 성별, 흡연여부를 tip의 평균
tips.groupby(['sex', 'smoker']).tip.mean()
```

## 그룹별 2개 이상의 통계 함수 적용

```python
# 성별 팁의 평균, 최댓값, 중앙값, 최솟값
tips.groupby('sex').tip.agg([np.mean, np.min, np.median, np.max]) # aggregate
```

```python
# 성별, 흡연여부 별 영수증과 팁의 평균과 표준편차
tips.groupby(['sex', 'smoker'])[['total_bill', 'tip']].agg([np.mean, np.std])
```

# `pivot_table` 함수(메서드)

*  Parameters:
    * values: 통계 함수를 적용할 변수(컬럼)
    * index: 결과 데이터프레임에서 행 인덱스로 사용할 변수(컬럼)
    * columns: 결과 데이터프레임에서 열 인덱스로 사용할 변수(컬럼)
    * aggfunc: aggregation function. 기본값은 평균

```python
# 성별 팁의 평균
tips.pivot_table(values='tip', index='sex') # aggfunc='mean' (기본값)
```

```python
# 성별 흡연여부별 팁의 평균
tips.pivot_table(values='tip', index=['sex', 'smoker']) # aggfunc='mean' (기본값)
```

```python
tips.pivot_table(values='tip', index='sex', columns='smoker')
```

```python
# 성별, 흡연여부별, 요일별, 시간별 팁의 평균
# 성별, 흡연여부 - 인덱스. 요일, 시간 - 컬럼
tips.pivot_table(values='tip', index=['sex', 'smoker'], columns=['day', 'time'])
```

```python
# 요일별, 시간별 영수증과 팁의 평균, 표준편차
tips.pivot_table(values=['total_bills', 'tip'],
                 index='day',
                 columns='times',
                 aggfunc=[np.mean, np.std])
```
