```python
import numpy as np  # ndarray
import pandas as pd  # Series, DataFrame
import matplotlib.pyplot as plt  # 시각화(그래프)
import seaborn as sns  # 시각화, 예제 데이터
```

# 데이터 셋 준비

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()   # 데이터프레임의 정보 - 타입, 행 인덱스, 컬럼 정보
```

matplotlib.pyplot 모듈을 사용한 시각화

1.  `figure` 객체 생성
1.  `figure` 객체에 subplot(axis)을 생성해서 추가
1.  subplot에 그래프 그림
1.  그래프의 여러가지 옵션들을 설정.
1.  그래프 보여주기.

```python
# 1. figure 객체 생성
figure = plt.figure()

# 2. 생성된 figure 객체에 subplot(axis)을 추가
# axis = figure.add_subplot(1, 1, 1)  #row, column, index
axis = figure.add_subplot(211)
figure.add_subplot(212)

# 3. subplot에 그래프 종류를 선택해서 그림
axis.hist(x=tips['total_bill'])

# 4. 그래프의 옵션 설정
axis.set_xlabel('total_bill')
axis.set_ylabel('Count')
axis.set_title('Histogram')
axis.grid()

# 5. 그래프 보여주기
plt.show()
```

```python
# 1. figure 생성
fig = plt.figure(figsize=(6, 8))

# 2. subplot을 figure에 추가
ax1 = fig.add_subplot(2, 1, 1)  # add_subplot(211)
ax2 = fig.add_subplot(2, 1, 2)  # add_subplot(212)

# 3. 그래프
ax1.hist(x=tips.total_bill)
ax2.scatter(x=tips.total_bill, y=tips.tip)

# 4. 옵션
ax1.grid()
ax2.grid()

# 5. 보여주기
plt.show()
```

figure에 2개의 subplot 좌우로 배치하고 hist, scatter 그래프를 그리기

```python
fig = plt.figure(figsize=(12, 4))

ax1 = fig.add_subplot(1, 2, 1) # add_subplot(121)
ax2 = fig.add_subplot(1, 2, 2) # add_subplot(122)

ax1.hist(x=tips.tip)
ax2.scatter(x=tips.total_bill, y=tips.tip)

ax1.grid
ax1.set_xlabel('Tip')
ax1.set_ylabel('Count')
ax1.set_title('Histogram')

ax2.grid()
ax2.set_xlabel('Total bill')
ax2.set_ylabel('Tip')
ax2.set_title('Histogram')

plt.show()
```

# `plt.subplots()` 함수

figure 생성 + subplot()

# figure에 1개의 subplot이 있는 경우

```python
# figure 생성, figure에 subplot 추가.
fig, ax = plt.subplots()

# subplot에 그래프 그림.
ax.hist(x=tips.total_bill)

# 옵션

#그래프 보여주기
plt.show()
```

## subplot 2개를 상하로 배치

```python
fig, ax = plt.subplots(nrows=2, figsize=(6, 8))
# print(ax)  # subplot이 2개 이상인 경우 리턴 값 ax는 subplot들의 배열(ndarray)

ax[0].hist(x=tips.total_bill)
ax[1].scatter(x=tips.total_bill, y=tips.tip)

plt.show()
```

## 2개의 subplot를 좌우로 배치

```python
fig, ax = plt.subplots(ncols=2, figsize=(12, 4))
# print(ax)

ax[0].hist(x=tips.total_bill)
ax[1].scatter(x=tips.total_bill, y=tips.tip)

plt.show()
```

# pyplot 모듈의 top-level 함수
 figure 생성 + subplot 1개 생성 + 그래프 종류

```python
# 1. figure 생성. 2. subplot 1개 생성. 3. histogram 그림.
plt.hist(x=tips.total_bill)

#4. 옵션 설정
plt.xlabel('Total bill')
plt.ylabel('Count')
plt.title('Histogram')

#5. 보여주기
plt.show()
```

## Histogram

연속형 변수를 구간으로 나눠서, 그 구간 안에 포함된 자료의 개수를 막대의 높이로 표현한 그래프.

```python
# 1. figure 생성.  2. subplot 1개 생성.  3. histogram 그림.
plt.hist(x=tips.total_bill, color='lightgray', edgecolor='black', bins=10)

# 4. 옵션 설정
plt.xlabel('Total bill')
plt.ylabel('Count')
plt.title('Histogram')
plt.grid()

# 5. 보여주기
plt.show()
```

hist 함수 파라미터:

* x: x축에 사용할 변수 이름. 히스토그램을 그리기 위한 자료.
* bins: 막대 개수, 구간
* color: 막대 안쪽을 채우는 색깔
* edgecolor: 막대 테두리 색깔
* density: False - 개수, True - 밀도(비율)

## Bar plot(막대 그래프)

카테고리 타입 변수의 빈도수를 막대로 표현한 그래프

```python
gender_count = tips.sex.value_counts()
gender_count
```

```python
type(gender_count)
```

```python
gender_count.index
```

```python
gender_count.values
```

```python
#세로막대그래프
plt.bar(x=gender_count.index, height=gender_count.values)

plt.show()
```

```python
# 가로 막대 그래프(horizontal graph)
plt.barh(y=gender_count.index, width=gender_count.values, height=0.5)

plt.show()
```
