## `barplot()` 함수

* 그룹(카테고리) 별로 변수의 통계량(평균, 표준편차, 중앙값, ...)을 막대 그래프로 표현.
* 통계량 계산 함수의 기본값은 평균(mean).
* 카테고리 타입 변수와 숫자 타입 변수를 함께 시각화.

```python
# 성별 tip의 평균
tips.groupby('sex').tip.mean()
```

```python
sns.barplot(data=tips, x='sex', y='tip')  #estimation = 'mean': 기본값, 평균 계산.
```

```python
tips.groupby('sex').tip.median()   # 중위값, 중앙값
```

```python
sns.barplot(data=tips, x='sex', y='tip', estimator='median')
plt.show()
```

```python
# 흡연여부별 tip과 total_bill의 평균을 시각화
sns.barplot(data=tips, x='smoker', y='tip')
plt.show()
```

```python
#  성별 흡연여부별 tip의 평균을 시각화
tips.groupby(['sex', 'smoker']).tip.mean()
```

```python
sns.barplot(data=tips, x='sex', y='tip', hue='smoker')
plt.show()
```

# 산점도: `scatterplot()`

* tip ~ total_bill 산점도

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)

plt.show()
```

* tip ~ total_bill
* 성별에 따라서 점의 색깔이 다르게 표현

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)

plt.show()
```

* tip ~ total_bill
* 성별에 따라서 점의 색깔을 다르게.
* 흡연여부에 따라서 점의 모양을 다르게.

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', data=tips)
```

* tip ~ total_bill 산점도
* 점의 색깔은 성별에 따라서 다르게 표현
* 점의 모양은 흡연여부에 따라서 다르게 표현
* 점의 크기를 인원수에 따라서 다르게 표현

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', sizes=(50, 200), data=tips)
```

# 카테고리 그래프: `catplot()`

카테고리 별로 subplot들을 생성하고 통계값(평균, 중앙값, ...) 계산해서, 각각의 subplot에 같은 종류의 그래프들을 시각화.

```python
# 요일별 팁의 평균을 막대 그래프로 표현
sns.barplot(data=tips, x='day', y='tip', hue='smoker')
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='tip', col='smoker')
plt.show()
```

* 요일별 팁의 평균 막대 그래프
* 시간(Lunch, Dinner) 별로 subplot을 생성.

```python
# Figure와 Subplot 생성
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Lunch 시간대의 요일별 팁의 평균 막대 그래프
sns.barplot(x='day', y='tip', data=tips[tips['time'] == 'Lunch'], ax=axes[0])

# Dinner 시간대의 요일별 팁의 평균 막대 그래프
sns.barplot(x='day', y='tip', data=tips[tips['time'] == 'Dinner'], ax=axes[1])

# 그래프 간격 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```

* 요일별 팁의 평균 막대 그래프
* 흡연여부별로 subplot
* 시간별 subplot

```python
# Figure와 Subplot 생성
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 요일별 팁의 평균 막대 그래프
sns.barplot(x='day', y='tip', data=tips, ax=axes[0, 0])

# 흡연여부별 요일별 팁의 평균 막대 그래프
sns.barplot(x='day', y='tip', hue='smoker', data=tips, ax=axes[0, 1])

# Lunch 시간대의 요일별 팁의 평균 막대 그래프
sns.barplot(x='day', y='tip', data=tips[tips['time'] == 'Lunch'], ax=axes[1, 0])

# Dinner 시간대의 요일별 팁의 평균 막대 그래프
sns.barplot(x='day', y='tip', data=tips[tips['time'] == 'Dinner'], ax=axes[1, 1])

# 그래프 간격 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```

* 요일별 성별 팁의 막대 그래프
* 시간별 subplot

```python
# Figure와 Subplot 생성
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Lunch 시간대의 요일별 성별 팁의 평균 막대 그래프
sns.barplot(x='day', y='tip', hue='sex', data=tips[tips['time'] == 'Lunch'], ax=axes[0])

# Dinner 시간대의 요일별 성별 팁의 평균 막대 그래프
sns.barplot(x='day', y='tip', hue='sex', data=tips[tips['time'] == 'Dinner'], ax=axes[1])

# 그래프 간격 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```

* 성별 팁의 box plot
* 요일별로 subplot

```python
# Figure와 Subplot 생성
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 요일별 성별 팁의 box plot
sns.boxplot(x='day', y='tip', hue='sex', data=tips, ax=axes[0, 0])

# 요일별 성별 팁의 분포와 바이올린 플롯
sns.violinplot(x='day', y='tip', hue='sex', split=True, data=tips, ax=axes[0, 1])

# Lunch 시간대의 요일별 성별 팁의 box plot
sns.boxplot(x='day', y='tip', hue='sex', data=tips[tips['time'] == 'Lunch'], ax=axes[1, 0])

# Dinner 시간대의 요일별 성별 팁의 box plot
sns.boxplot(x='day', y='tip', hue='sex', data=tips[tips['time'] == 'Dinner'], ax=axes[1, 1])

# 그래프 간격 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```
