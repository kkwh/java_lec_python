# 패키지 임포트

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# mpg 데이터셋

```python
mpg = sns.load_dataset('mpg')

mpg.head()
```

```python
mpg.info()
```

mpg(Mile Per Gallon) dataset: 자동차 연비 데이터 셋

* mpg: 연비
* cylinders: 실린더 개수
* displacement: 배기량
* horsepower: 마력
* weight: 자동차 무게
* acceleration: 가속도
* model_year: 자동차 생산 연도
* origin: 자동차 생산지
* name: 자동차 이름

* mpg 데이터 셋 탐색.
* mpg ~ displacement scatter plot
* 2x2 subplot을 생성. scatter plot.
    * mpg ~ cylinders
    * mpg ~ horsepower
    * mpg ~ weight
    * mpg ~ accleration
* cylinders 별 mpg 평균 막대 그래프
* cylinders 별 mpg box plot
* origin별, cylinders 별 mpg box plot
* origin별, cylinders 별 mpg 평균 막대 그래프

```python
# mpg와 displacement 간의 scatter plot
sns.scatterplot(x='displacement', y='mpg', data=mpg)
plt.title('mpg ~ displacement Scatter Plot')
plt.xlabel('Displacement')
plt.ylabel('MPG')
plt.show()
```

```python
# variables = ['cylinders', 'horsepower', 'weight', 'acceleration'] # 독립변수

# # 2x2 sublot들을 생성:
# fig, a
```

```python
# Figure와 Subplot 생성
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# mpg ~ cylinders scatter plot
sns.scatterplot(x='cylinders', y='mpg', data=mpg, ax=axes[0, 0])
axes[0, 0].set_title('mpg ~ cylinders Scatter Plot')
axes[0, 0].set_xlabel('Cylinders')
axes[0, 0].set_ylabel('MPG')

# mpg ~ horsepower scatter plot
sns.scatterplot(x='horsepower', y='mpg', data=mpg, ax=axes[0, 1])
axes[0, 1].set_title('mpg ~ horsepower Scatter Plot')
axes[0, 1].set_xlabel('Horsepower')
axes[0, 1].set_ylabel('MPG')

# mpg ~ weight scatter plot
sns.scatterplot(x='weight', y='mpg', data=mpg, ax=axes[1, 0])
axes[1, 0].set_title('mpg ~ weight Scatter Plot')
axes[1, 0].set_xlabel('Weight')
axes[1, 0].set_ylabel('MPG')

# mpg ~ acceleration scatter plot
sns.scatterplot(x='acceleration', y='mpg', data=mpg, ax=axes[1, 1])
axes[1, 1].set_title('mpg ~ acceleration Scatter Plot')
axes[1, 1].set_xlabel('Acceleration')
axes[1, 1].set_ylabel('MPG')

# 그래프 간격 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
# cylinders 별 mpg의 평균 막대 그래프
sns.barplot(x='cylinders', y='mpg', data=mpg)

plt.xlabel('Cylinders')
plt.ylabel('MPG')
plt.show()
```

```python
# cylinders 별 mpg의 box plot
sns.boxplot(x='cylinders', y='mpg', data=mpg)

plt.xlabel('Cylinders')
plt.ylabel('MPG')
plt.show()
```

```python
# origin별, cylinders 별 mpg의 box plot
sns.boxplot(x='origin', y='mpg', hue='cylinders', data=mpg)

plt.xlabel('Origin')
plt.ylabel('MPG')
plt.show()
```

```python
# origin별, cylinders 별 mpg의 평균 막대 그래프
sns.barplot(x='origin', y='mpg', hue='cylinders', data=mpg)

plt.xlabel('Origin')
plt.ylabel('MPG')
plt.show()
```
